**Title**: Clustering of mall customers

**Context**: You are a data scientist working for a shopping mall that wants to understand the behavior of its customers to improve their shopping experience. To do this, you have access to a database containing information about customers such as their age, annual income, and annual spending on purchases.

**Objective**: Use the agglomerative clustering algorithm with the complete linkage method to group customers into different segments based on their demographic characteristics and shopping habits. Then, visualize the different clusters using a color palette.

**Data**: The [data](https://www.kaggle.com/datasets/kondapuramshivani/mall-customerscsv) is stored in a CSV file named "Mall_Customers.csv" containing 200 rows and 5 columns. The columns are "CustomerID", "Gender", "Age", "Annual Income (in thousands of dollars)", and "Spending Score (1-100)".

**Instructions**:

1. Load the data from the CSV file and store the data in a variable named "data".

2. Select the columns "Age", "Annual Income (in thousands of dollars)", and "Spending Score (1-100)" using the .iloc() method of Pandas and store the result in a variable named "X".

3. Use the agglomerative clustering algorithm with the complete linkage method to group customers into different segments. To do this, import the "AgglomerativeClustering" class from the scikit-learn library and initialize it with the parameters "n_clusters=5" and "affinity='euclidean'".

4. Fit the model with the "X" data and store the result in a variable named "clusters".

5. Display the dendrogram using the "dendrogram" function from the scipy.cluster.hierarchy library.

6. Use the "scatter" function from the matplotlib library to visualize the different clusters. To do this, use the "Annual Income (in thousands of dollars)" and "Spending Score (1-100)" columns of the "X" variable to create the graph and color each point according to its assigned cluster.

```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# Loading the dataset
dataset = pd.read_csv('/content/Mall_Customers .csv')

# Selecting the features
X = dataset.iloc[:, [3, 4]].values

# Using complete linkage for distance calculation and creating the dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='complete'))

# Creating the clusters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'complete')
y_hc = hc.fit_predict(X)
plt.show()

# Visualizing the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

This code performs hierarchical clustering on a dataset of mall customers in order to group them into different clusters based on their annual income and spending score.

The first step is to import the necessary libraries: pandas for data manipulation, matplotlib for visualization, and scipy.cluster.hierarchy for hierarchical clustering.

Next, the dataset is loaded from a CSV file named "Mall_Customers.csv" and the two relevant features, "Annual Income" and "Spending Score", are selected and stored in a variable named "X".

The dendrogram is created using complete linkage as the method for distance calculation. The dendrogram is a visual representation of the hierarchy of clusters and can be used to determine the appropriate number of clusters to use.

The next step is to create the clusters using agglomerative clustering with five clusters, Euclidean distance as the measure of similarity, and complete linkage as the method for merging clusters. The resulting clusters are stored in a variable named "y_hc".

Finally, the clusters are visualized using a scatter plot, with each cluster represented by a different color. The x-axis shows annual income, and the y-axis shows spending score. The title of the plot is "Clusters of customers".





**Cluster 1 (Red)**: High income, low spending score. These customers have high purchasing power but are not spending much at the mall, so they could be targeted with specific offers to increase their spending.

**Cluster 2 (Blue)**: Average income, average spending score. These customers are neither high nor low on any factor, so they are a neutral group that may require additional analysis to understand their behavior.

**Cluster 3 (Green)**: Low income, high spending score. These customers have a low purchasing power but are spending a lot at the mall, so they may be attracted by lower-priced products or promotions.

**Cluster 4 (Cyan)**: Low income, low spending score. These customers have limited purchasing power and are not spending much at the mall, so they may require special deals or discounts to incentivize them to make purchases.

Cluster 5 (Magenta) : High income, high spending score. These customers have high purchasing power and are also spending a lot at the mall, making them a valuable target group for marketing efforts.
