**I. Unsupervised Association Rules**

**I.1. Apriori Concept**

```python
import mlxtend # Importation de la bibliothèque mlxtend
import pandas as pd
```

```python
# Création d'une base de données
data = [['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
        ['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
        ['Milk','Apple','Kidney Beans','Eggs'],
        ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
        ['Corn','Onion','Onion','Kidney Beans','Ice cream','Eggs']]
```

```python
# Transformation de la base de données des transaction à une base de données exploitable.
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(data).transform(data) 
te_ary
df=pd.DataFrame(te_ary, columns=te.columns_) 
df
```

**I.1.2. Support**

La métrique support ou la fréquence d’apparition des produits et ensembles de produits dans un panier.

```python
from mlxtend.frequent_patterns import apriori 
```

```python
f_i = apriori(df, min_support = 0.6, use_colnames= True)
f_i
```

**I.1.3. confidence**

La probabilité d'acheter un produit X après avoir acheter Y


```python
from mlxtend.frequent_patterns import association_rules 
association_rules(f_i,metric="confidence",min_threshold=0.7) 
```

**I.1.4. Lift**

Quelle est la probabilité d'acheter les articles X et Y ensemble plutôt que d'acheter uniquement l'article X


```python
from mlxtend.frequent_patterns import association_rules 
association_rules(f_i,metric="lift",min_threshold=1.25)
```
