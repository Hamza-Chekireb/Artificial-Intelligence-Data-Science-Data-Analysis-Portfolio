#1.Classification(Classification des fleurs Iris)

##1.1. Importation des bibliothèques et fonctions

```python
#0. Importation des bibliothèques
import pandas as pd
import numpy as np

# Importation d'un ensemble de base de données téléchargeable
from sklearn import datasets

# Importation de TensorFlow pour le deep learning
import tensorflow as tf

# Importation des fonctions pour le preprocessing
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Importation de la fonction pour la division des données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
```

##1.2. Data Preprocessing

```python
# 1. Chargement du jeu de données
iris = datasets.load_iris()

# 2. Séparation des caractéristiques et de la variable cible
# Caractéristiques
X = iris.data[:, :]
# Variable cible
y = iris.target

# Conversion en one-hot encoded en utilisant tf.keras
Y = tf.keras.utils.to_categorical(y)

# 3. Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1)
```

##1.3.Construction du modèle

```python
# Importation des bibliothèques
import tensorflow as tf
# Architecture de couches en parallèle
from tensorflow.keras.models import Sequential
# Touts les neurones des couches précédentes sont
# Connectés à tous les neurones des couches suivantes
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# 4. Construction du modèle
# Réseau de neurones artificiels
# Initialisation du modèle
model = Sequential([
   Dense(16, input_dim=4, activation='relu'),
   Dense(3, activation='softmax')
])
```

##1.4. Entrainement du modèle

```python
# 5. Entraînement et Ajustement du modèle

# Configuration du modèle (configuration de la rétropropagation du gradient)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=["accuracy"])

import time # Calculer le temps d’entraînement

start = time.time()

history = model.fit( X_train, y_train, validation_split = 0.2,  epochs= 300)

print("The time used to execute this is given below")

end = time.time()

print(end - start)
```

##1.5. Voir l'architecture du modèle

```python
# Le sommaire

model.summary()
```

##1.6. Historique d’entraînement et évaluation

```python
# l'historique de l'Accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='lower right')
plt.show()
```

```python
# l'historique du coût 'loss'
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
